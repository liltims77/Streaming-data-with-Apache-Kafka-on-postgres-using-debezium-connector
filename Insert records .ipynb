```python
#import needed libraries
from sqlalchemy import create_engine
import pandas as pd
import os
import time

engine = create_engine(f'postgresql://postgres:<your-password>@localhost:5432/uber_latest')
```

```python
df = pd.read_sql('SELECT * FROM public.yellow_trip_data', engine)
df = df[['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'passenger_count', 'trip_distance', 'store_and_fwd_flag', 'fare_amount']]
df.head()
```

```python
for index, row in df.head(30).iterrows():
    mod = pd.DataFrame(row.to_frame().T)
    mod.to_sql(f"yellow_trip_data_streaming", engine, if_exists='append', index=False)
    print("Row Inserted " + mod.passenger_count.astype(str) + ' ' + mod.fare_amount.astype(str).astype(str))
    time.sleep(3)
```
